```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\ATHUL AKSHAY\Desktop\Machine Learning Projects\SQL\demand_gendata.csv")
data.head(5)
```

```python
data.groupby(['month','Product']).sum()
```

```python
profitable_demand=pd.DataFrame(data.groupby(['month','Product']).sum()['demand'])
profitable_demand
```

```python
profitable_demand.sort_values(by=['month','demand'],inplace=True,ascending=[True,False])
profitable_demand
```

```python
profitable_demand.reset_index(inplace=True)
profitable_demand
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,7.5))
sns.barplot(x='Product',y='demand',hue='month',data=profitable_demand)
plt.title('Highly Demanded Products per month')
plt.xlabel('Products')
plt.ylabel('Demand')
plt.show()
```

```python
def sales_per_month(month_selected):
    data_selected=data[data['month']==month_selected]
    # print(data_selected)
    plt.figure(figsize=(15,7))
    sns.barplot(x='Product',y='demand',data=data_selected,palette='viridis',color='blue')
    plt.title(f'Sales Details for {month_selected}')
    plt.xlabel('Products')
    plt.ylabel('Demand')
    highly_sold_product=data_selected.loc[data_selected['demand'].idxmax()]['Product']
    less_sold_product=data_selected.loc[data_selected['demand'].idxmin()]['Product']

    print(f'Highly Sold product for {month_selected}:{highly_sold_product}')
    print(f'Less Sold product for {month_selected}:{less_sold_product}')
sales_per_month('February')
```
